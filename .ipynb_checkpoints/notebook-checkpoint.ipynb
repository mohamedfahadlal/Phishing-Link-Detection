```python
from scipy.io import arff
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

data, meta = arff.loadarff(r"C:\Users\Fahad\Desktop\my works\IML\Project\Phishing-Link-Detection\Data\Training Dataset.arff")


df = pd.DataFrame(data)

print(df.head())  

```

```python
df_clean = df.applymap(lambda x: int(x.decode('utf-8')) if isinstance(x, bytes) else x)
```

```python
df_clean['Result'].value_counts(normalize=True)
```
